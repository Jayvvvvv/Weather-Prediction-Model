```python
# Weather['Location'].fillna(Weather['Location'].mode(), inplace=True)
# Weather['WindGustDir'].fillna(Weather['WindGustDir'].mode(), inplace=True)
# mode = Weather['WindDir9am','WindDir3pm'].mode()

# mode1 = Weather['WindGustDir'].mode()
# print(mode)
# print(mode1)

Weather.drop(columns=['Date'], axis=1, inplace=True)
Weather
```

```python
from sklearn.preprocessing import LabelEncoder
ln = LabelEncoder()
Weather[['Location','WindDir9am', 'WindDir3pm', 'RainToday', 'RainTomorrow', 'WindGustDir']] = Weather[['Location', 'WindDir9am', 'WindDir3pm', 'RainToday', 'RainTomorrow', 'WindGustDir']].apply(ln.fit_transform)
```

```python
Weather.fillna(Weather.mode(), inplace=True)
Weather.fillna(Weather.median(), inplace=True)
```

```python
Weather
```

```python
Weather.isnull().sum()
```

```python
x = Weather.drop('RainTomorrow', axis=1)
y = Weather['RainTomorrow']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.5)
```

```python
print(x_train.shape, x_test.shape)
print(y_train.shape, y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(x_train,y_train)
```

```python
pred = knn.predict(x_test)
pred
```

```python
from sklearn.metrics import accuracy_score
KNN = round(accuracy_score(y_test, pred)*100,2)
print(KNN)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(x_train, y_train)
```

```python
pred1 = dtc.predict(x_test)
pred1
```

```python
DT = round(accuracy_score(y_test, pred1)*100,2)
print(DT)
```

```python
from sklearn.linear_model import LogisticRegression
lgr = LogisticRegression()
lgr.fit(x_train, y_train)
```

```python
pred2 = lgr.predict(x_test)
pred2
```

```python
LR = round(accuracy_score(y_test, pred2)*100,2)
print(LR)
```

```python
from sklearn.svm import SVC
Svm = SVC()
Svm.fit(x_train, y_train)
```

```python
pred3 = Svm.predict(x_test)
pred3
```

```python
SVM = round(accuracy_score(y_test, pred3)*100,2)
print(SVM)
```
